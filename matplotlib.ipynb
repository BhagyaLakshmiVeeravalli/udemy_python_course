```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [1500, 1200, 1100, 1800]
```

```python
plt.plot(x, y)
```

```python
legend = ['January', 'February', 'March', 'April']
```

```python
plt.xticks(x, legend)
plt.plot(x, y)
plt.show()
```

```python
plt.bar(x, y)
plt.ylabel('Sales in US$')
plt.title('Monthly sale')
```
